```python
import matplotlib.pyplot as plt
data = pd.read_csv('Weather_Data[1].csv')
x_data = data['Date/Time']
y_data = data['Temp_C']

plt.plot(x_data, y_data)
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Temperature on January 1st and 3rd')
plt.show()

```

```python
print(data.columns)
weathercounts = data['Weather'].value_counts(ascending=True)[-20:]
weathercounts.plot(kind='barh', figsize=(12, 5))
plt.title('Frequency of Different Weather Conditions')
plt.xlabel('Frequency')
plt.ylabel('Weather Condition')
plt.show()
```

```python
import seaborn as sns
print(data.columns)
seasoncolumn = 'Weather'
weathercolumn = 'Temp_C'
groupeddata = data.groupby([seasoncolumn, weathercolumn]).size().reset_index(name='Count')
topconditions = groupeddata.groupby(weathercolumn)['Count'].sum().nlargest(10).index
filtereddata = groupeddata[groupeddata[weathercolumn].isin(topconditions)]
plt.figure(figsize=(12, 6))
sns.barplot(data=filtereddata, x=weathercolumn, y='Count', hue=seasoncolumn, ci=None)
plt.title('Frequency of Different Weather Conditions by Season')
plt.xlabel('Frequency')
plt.ylabel('Weather Condition')
plt.legend(title='Season')
plt.show()
```

```python
plt.hist(data['Temp_C'], bins=50)
plt.xlabel('Date/Time')
plt.ylabel('Temp_C')
plt.title('Temperature')
plt.show()
```

```python
category_counts = data['Weather'].value_counts()
plt.bar(category_counts.index, category_counts.values)
plt.xlabel('Weather')
plt.ylabel('Date/Time')
plt.title("weather bar")
plt.xticks(rotation=45)
plt.show()
```

```python
wind_attr_counts = data['Wind Speed_km/h'].value_counts()
plt.pie(wind_attr_counts, labels=wind_attr_counts.index, autopct='%.1f%%')
plt.title('wind speed')
plt.show()
```

```python
data = {'Weather': ['Cloudy', 'Sunny', 'Rainy', 'Cloudy', 'Sunny', 'Sunny', 'Rainy', 'Foggy']}
dataset = pd.DataFrame(data)
top3weather = dataset['Weather'].value_counts().nlargest(3)
proportions = top3weather / top3weather.sum()
plt.pie(proportions, labels=proportions.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 3 Weather Conditions')
plt.show()
```
